```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
def RGBaIYQ(img):
    imgAux=imageio.imread(img)/255
    w=imgAux.shape[0]
    h=imgAux.shape[1]
    p=3
    m_yiq=np.zeros((w,h,p))
    m_yiq.shape
    m_yiq[:,:,0]=0.299*imgAux[:,:,0]+0.587*imgAux[:,:,1]+0.114*imgAux[:,:,2]
    m_yiq[:,:,1]=0.595716*imgAux[:,:,0]+-0.274453*imgAux[:,:,1]+-0.321263*imgAux[:,:,2]
    m_yiq[:,:,2]=0.211456*imgAux[:,:,0]+-0.522591*imgAux[:,:,1]+-0.321263*imgAux[:,:,2]
    
    return m_yiq
    
    
```

```python
im=RGBaIYQ('lena128C.png')
im=np.clip(im/1.,0.,1.)
#print(im)
print(im.shape,im.dtype)
plt.imshow(im)
plt.figure()
plt.show()
```

```python
def prodxescalar(imagen):
    w=imagen.shape[0]
    h=imagen.shape[1]
    p=3
    yiqbis=np.zeros((w,h,p))
    yiqbis.shape
    yiqbis[:,:,0]=0.5*imagen[:,:,0]
    yiqbis[:,:,1]=0.2*imagen[:,:,1]
    yiqbis[:,:,2]=0.2*imagen[:,:,2]
    
    return yiqbis
```

```python
var1=prodxescalar(im)
#print(var1)
print(im.shape,var1.dtype)
plt.imshow(var1)
plt.figure()
plt.show()
```

```python
def IYQaRGB(img):
    #imgAux2=imageio.imread(img)
    w=img.shape[0]
    h=img.shape[1]
    p=3
    m_rgb=np.zeros((w,h,p))
    m_rgb.shape
    m_rgb[:,:,0]=1*img[:,:,0]+0.9663*img[:,:,1]+0.6210*img[:,:,2]
    m_rgb[:,:,1]=1*img[:,:,0]+-0.2721*img[:,:,1]+-0.6474*img[:,:,2]
    m_rgb[:,:,2]=1*img[:,:,0]+-1.1070*img[:,:,1]+-1.7046*img[:,:,2]
    
    return m_rgb
```

```python
im2=IYQaRGB(var1)
im2 = np.clip(im2/1.,0.,1.)
#print(im2)
print(im2.shape,im2.dtype)
plt.imshow(im2)
plt.figure()
plt.show()
```
